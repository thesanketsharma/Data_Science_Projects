## Introduction: 

Fraud is prevalent everywhere, be it Payments, E-commerce, Healthcare or banking systems. Fraud is a billion-dollar business and even increasing. About 49% of companies have experienced fraud of one or another kind. 

Fraud might seem scary for people and businesses, but it can be detected using intelligent systems like rule engines and machine learning. A rules engine is a software system that executes one or more business rules written by domain experts in a runtime production environment for detecting fraudulent transactions. One Problem with rules is that they can fire a lot of false positives or false negatives due to predefine threshold values. For tackling these kind of problems ML comes in handy. The combination of rule-engines and machine learning, can make fraud detection more accurate and precise.

## Objective:
Our objective is to build a machine learning model to Identify normal and fraudulent payment transactions.

In this notebook we will do:
1. Exploratory Data Analysis (EDA)
2. Data Preprocessing
3. Oversampling with SMOTE

5. Ensembling best performing models into a soft voting classifier
6. Conclusion


## About Dataset

Dataset has been sourced from Kaggle 
Dataset link: https://www.kaggle.com/datasets/ealaxi/banksim1/code?resource=download

This synthetically generated dataset consists payments data from various customers made in different time periods and with different amounts. For more information on the dataset you can check the Kaggle page for this dataset.

### Data:
Dataset Has 9 feature columns and 1 target column:
* Step: This feature represents the day from the start of simulation. It has 180 steps so simulation ran for virtually 6 months.
* Customer: This feature represents the customer id
* zipCodeOrigin: The zip code of origin/source.
* Merchant: The merchant's id
* zipMerchant: The merchant's zip code
* Age: Categorized age
        0: <= 18,
        1: 19-25,
        2: 26-35,
        3: 36-45,
        4: 46:55,
        5: 56:65,
        6: > 65
        U: Unknown
* Gender: Gender for customer
        E : Enterprise,
        F: Female,
        M: Male,
        U: Unknown
* Category: Category of the purchase. 
* Amount: Amount of the purchase
* Fraud: Target variable which shows if the transaction fraudulent(1) or benign(0)
