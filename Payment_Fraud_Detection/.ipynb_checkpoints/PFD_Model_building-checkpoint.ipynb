# Payment Fraud Detetcion

```python
# importing necessary libraries

# For Data Analysis & visualisations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()   # setting seaborn style

#Pipeline
from sklearn.pipeline import Pipeline

# For SMOTE (synthetic minority oversampling technique)
from imblearn.over_sampling import SMOTE

#Metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, roc_auc_score

# Classification Models
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Loading Data Set
data = pd.read_csv("Data/Banksim.csv")
data.head()
```

```python
#selecting 25% of data
datasample = data.sample(frac=0.25, axis=0, random_state=0 )
```

```python
#input and output features
X= datasample.drop(labels='fraud', axis=1) # input features
y = datasample['fraud'] # output
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### Preprocessing Data

```python
# function  for preprocessing
def preprocess(df, useless_columns):
    #dropping useless columns
    df1 = df.drop(labels= useless_columns, axis=1)
    
    # droping columns having more than 80% nan values
    df2 = df1.drop(df1.columns[df1.apply(lambda col: col.isnull().sum()/len(df) > 0.8)], axis=1)
    
    # filling nan values by zero
    df3 = df2.fillna(0)
    
    # finding categorial columns
    object_columns = [i for i in df3.select_dtypes(include=['object']).columns]
    
    # one hot encoding for categorial columns
    df4 = pd.get_dummies(df3, columns = object_columns)
    return df4
    
```

```python
useless_columns = ['step', 'customer', 'zipcodeOri', 'merchant', 'zipMerchant']
# preprocessing training data
X_train = preprocess(X_train, useless_columns)

#preprocessing test data
X_test = preprocess(X_test, useless_columns)
```

### Oversampling (SMOTE)

```python
oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

### Model Building

```python
# Grid search Cross validation for selecting best parameters
grid = { 
    'n_estimators': [50, 100, 200],
    'criterion': ["gini", "entropy"],
    'max_depth' : [5,10,15,20]
}

model = RandomForestClassifier(random_state=0, verbose=1, n_jobs=-1)

rf_cv = GridSearchCV(estimator=model, param_grid=grid, cv=5)

rf_cv.fit(X_train, y_train)

rf_cv.best_params_
```

```python
# hyperparameter tuning round 2
grid = {'max_depth' : [16,17,18,19,20]}

model = RandomForestClassifier(criterion='entropy', max_depth = 18, n_estimators=50, random_state=0, verbose=1, n_jobs=-1)

rf_cv = GridSearchCV(estimator=model, param_grid=grid, cv=5)

rf_cv.fit(X_train, y_train)

rf_cv.best_params_
```

```python
# hyperparameter tuning round 3
grid = {'n_estimators' : [51,52,53,54,55,56]}

model = RandomForestClassifier(criterion='entropy', max_depth = 18, random_state=0, verbose=1, n_jobs=-1)

rf_cv = GridSearchCV(estimator=model, param_grid=grid, cv=5)

rf_cv.fit(X_train, y_train)

rf_cv.best_params_
```

```python
# Final Model
rf_model = RandomForestClassifier(criterion='entropy', max_depth = 18, n_estimators=54, random_state=0, verbose=1, n_jobs=-1)
rf_model.fit(X_train, y_train)
```

```python
#predictions
pred_y_train= rf_model.predict(X_train)
pred_y_test = rf_model.predict(X_test)
```

```python
# Model evaluation Using classification report and ROC_AUC_SCORE
print("Classification Report on Training Data: \n", classification_report(y_train, pred_y_train))
print("Classification Report on Test Data: \n", classification_report(y_test, pred_y_test))

print("AUC_Score on Training Data:", roc_auc_score(y_train, pred_y_train))
print("AUC_Score on Test Data:", roc_auc_score(y_test, pred_y_test))
```

```python
confusion_matrix(y_test, pred_y_test)
```
